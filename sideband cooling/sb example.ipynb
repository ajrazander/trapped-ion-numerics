# Using the sideband_cooling.simulation module
The simulation module under sideband cooling is meant to be flexible enough to customize every single sideband cooling (SBC) pulse, but also automated enough to allow for simple cooling schemes.

There are 3 built in cooling protocols:
* Classic -- applies suscessive $\pi$-pulses associated with decreasing harmonic states; very fast computation, but poor cooling for nonsmall $\eta$ and $\bar{n}_i \gtrsim 10$
* Fixed -- applies the same pulse length for every pulse; fastish computation (has an optimization layer in simulation.py); good for looking nonsmall $\eta$ and $\bar{n}_i \gtrsim 10$
* Optimal -- global optimization that applies the optimal pulse length given $N$ pulses; slow computation after a couple dozen pulses (scales exponentially); best possible cooling per pulse for "any" $\eta$ and "any" $\bar{n}_i$

Any order RSB can be applied at any pulse time. Additional optimization outside of simulation.py is required to do that. (See multiorder cooling example below.)

The simulation module requires some experimental parameters outside the specified cooling protocol and number of cooling pulses such as the Rabi frequency $\Omega$, the initial average harmonic state $\bar{n}_i$, and the confined harmonic frequency along the direction of cooling $\omega_z$ in the example below (used to calculate $\eta$ assuming 355 nm Raman cooling).

```python
import numpy as np

import matplotlib.pyplot as plt

from tqdm import tqdm

from sideband_cooling import simulation
```

### Example using classic
Simple use case for classic protocol cooling with 40 pulses.

```python
# Example usage of simulation package

# define constants
omega_z = 2*np.pi * 1e6
initial_nbar = 19.6e6 / 2 / (omega_z / (2*np.pi))  # Doppler limited initial nbar for 171Yb+
omega = 2*np.pi/(2 * 5e-6)

sim = simulation.SidebandCooling(omega_z, initial_nbar, omega)

strategy = 'classic'
N_pulses = 40

cooled_nbar, pulse_schedule, cooling_time, cooled_dist = sim.cool(strategy, N_pulses, 1)
```

```python
# Plot results
print('initial nbar', initial_nbar, ' now cooled to', np.round(cooled_nbar, 3))

n_trunc = 100
ns_partial = np.arange(0, n_trunc)

plt.plot(ns_partial, cooled_dist[:n_trunc])
plt.xlabel(r'harmonic state ($n$)', fontsize=12)
plt.ylabel('probability', fontsize=12)
plt.title('cold ion harmonic distribution', fontsize=16)
plt.yscale('log')
plt.grid()
plt.show()
```

### Example using fixed
Simple use case for fixed protocol using 40 pulses.

```python
# Plot nbar as a function of number of pulses

# define constants
omega_z = 2*np.pi * 1e6
initial_nbar = 19.6e6 / 2 / (omega_z / (2*np.pi))
omega = 2*np.pi/(2 * 5e-6)

sim = simulation.SidebandCooling(omega_z, initial_nbar, omega)

strategy = 'fixed'
list_pulses = np.arange(0, 40)

cooled_nbars = []
pulse_schedules = []
cooling_times = []
cooled_dists = []
for N_pulses in tqdm(list_pulses):
    cooled_nbar, pulse_schedule, cooling_time, cooled_dist = sim.cool(strategy, N_pulses, 1)
    cooled_nbars.append(cooled_nbar)
    pulse_schedules.append(pulse_schedule)
    cooling_times.append(cooling_time)
    cooled_dists.append(cooled_dist)
    sim.reset_current_dist()
```

```python
# Plot results

plt.plot(list_pulses, cooled_nbars)
plt.xlabel('number of pulses', fontsize=12)
plt.ylabel(r'$\bar{n}$', fontsize=12)
plt.title(strategy+r' strategy with final $\bar{n}=$'+str(cooled_nbars[-1])[:4], fontsize=16)
plt.grid()
plt.show()
```

### Example using optimal

```python
# Plot nbar as a function of number of pulses

# define constants
omega_z = 2*np.pi * 1e6
initial_nbar = 19.6e6 / 2 / (omega_z / (2*np.pi))
omega = 2*np.pi/(2 * 5e-6)

sim = simulation.SidebandCooling(omega_z, initial_nbar, omega)

strategy = 'optimal'
list_pulses = np.arange(1, 40)

cooled_nbars = []
pulse_schedules = []
cooling_times = []
cooled_dists = []
for N_pulses in tqdm(list_pulses):
    cooled_nbar, pulse_schedule, cooling_time, cooled_dist = sim.cool(strategy, N_pulses, 1)
    cooled_nbars.append(cooled_nbar)
    pulse_schedules.append(pulse_schedule)
    cooling_times.append(cooling_time)
    cooled_dists.append(cooled_dist)
    sim.reset_current_dist()
```

```python
# Plot results

plt.plot(list_pulses, cooled_nbars)
plt.xlabel('number of pulses', fontsize=12)
plt.ylabel(r'$\bar{n}$', fontsize=12)
plt.title(strategy+r' strategy with final $\bar{n}=$'+str(cooled_nbars[-1])[:4], fontsize=16)
plt.grid()
plt.show()
```

### Estimating the best number of 1st and 2nd order pulses with optimal (2nd order first)

```python
# Apply mutli-order cooling

# define constants
omega_z = 2*np.pi * 1e6
initial_nbar = 19.6e6 / 2 / (omega_z / (2*np.pi))
omega = 2*np.pi/(2 * 5e-6)

sim = simulation.SidebandCooling(omega_z, initial_nbar, omega)

strategy = 'fixed'
pulses_1st_order = np.arange(2, 12)
pulses_2nd_order = pulses_1st_order[::-1]

cooled_nbars = []
pulse_schedules = []
cooling_times = []
cooled_dists = []
for p1s, p2s in tqdm(zip(pulses_1st_order, pulses_2nd_order), total=len(pulses_1st_order)):
    cooled_nbar2, pulse_schedule2, cooling_time2, cooled_dist2 = sim.cool(strategy, p2s, 2)
    cooled_nbar, pulse_schedule, cooling_time, cooled_dist = sim.cool(strategy, p1s, 1)
    cooled_nbars.append(cooled_nbar)
    pulse_schedules.append(pulse_schedule)
    cooling_times.append(cooling_time)
    cooled_dists.append(cooled_dist)
    sim.reset_current_dist()
```

```python
# Plot results

plt.plot(pulses_1st_order, cooled_nbars)
plt.xlabel('number of 1st order pulses', fontsize=12)
plt.ylabel(r'$\bar{n}$', fontsize=12)
plt.title(strategy+r' strategy with minimum $\bar{n}=$'+str(np.min(cooled_nbars))[:4], fontsize=16)
plt.grid()
plt.show()

plt.plot(pulses_2nd_order, cooled_nbars)
plt.xlabel('number of 2nd order pulses', fontsize=12)
plt.ylabel(r'$\bar{n}$', fontsize=12)
plt.title(strategy+r' strategy with minimum $\bar{n}=$'+str(np.min(cooled_nbars))[:4], fontsize=16)
plt.grid()
plt.show()
```
